```python
import pandas as pd

data = {
    'Name': ['Rahul', 'Priya', 'Amit', 'Neha'],
    'Age': [22, 21, 23, 22],
    'City': ['Pune', 'Mumbai', 'Nashik', 'Delhi'],
    'Marks': [78, 85, 90, 67],
    'Course': ['Python', 'Java', 'Python', 'Data Science']
}

df = pd.DataFrame(data)

df.to_excel("students.xlsx", index=False)

print("students.xlsx created successfully!")

df = pd.read_excel("students.xlsx")
print(df.head())

# First 5 rows
print(df.head())

# Shape
print(df.shape)

# Column names
print(df.columns)

# Data types
print(df.dtypes)
```

```python


data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Printer', 'Headphones'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories'],
    'Price': [50000, 20000, 15000, 8000, 2000],
    'Quantity': [5, 10, 8, 4, 15]
}

df1 = pd.DataFrame(data)

print(df1)
```

```python
print(df1[df1['Category'] == 'Electronics'])
```

```python
print(df1[df1['Price'] > 10000])
```

```python
print(df1['Price'].mean())
```

```python
print(df1[df1['Price'] == df1['Price'].max()])
```

```python
print(df1[df1['Price'] == df1['Price'].min()])
```

```python
print(df1[df1['Quantity'] > 7])
```

```python
df1['Total_Price'] = df1['Price'] * df1['Quantity']

print(df1)
```
